```python
from PIL import Image
import os

# Paths
input_folder = "/media/lars/F77E-7D07/thumbnails/"
output_folder = "/media/lars/F77E-7D07/thumbnails_resized/"
sizes = {"small": (100, 100), "medium": (250, 250), "full" : (400,400)}  # Target sizes

# Ensure output directories exist
for size in sizes.keys():
    os.makedirs(os.path.join(output_folder, size), exist_ok=True)

# Process all images in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith(".jpg") or filename.endswith(".png"):  # Support both JPG and PNG
        input_path = os.path.join(input_folder, filename)
        img = Image.open(input_path)

        # Convert RGBA to RGB (remove alpha channel)
        if img.mode == "RGBA":
            img = img.convert("RGB")

        for size_name, size in sizes.items():
            resized_img = img.resize(size, Image.LANCZOS)  # High-quality resizing
            output_path = os.path.join(output_folder, size_name, filename)
            resized_img.save(output_path, "JPEG", quality=90)  # Save with high quality

        print(f"Processed: {filename}")

print("✅ Image resizing completed!")
```
